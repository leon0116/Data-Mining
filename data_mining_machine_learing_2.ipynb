```python
dataset = pd.read_excel('health.xlsx')
dataset
```

```python
dataset['每單位卡路里之行走步數'] = dataset.iloc[:,1]/dataset.iloc[:,2]
dataset
```

```python
bins = [0,2.5,5,10]
label1 = ['低','中','高']
dataset['健康指數'] = pd.cut(dataset['每單位卡路里之行走步數'],bins,right=False,labels=label1)
dataset
```

```python
onehot_encoding=pd.get_dummies(dataset['健康指數'],prefix='健康指數')
pd.concat([dataset,onehot_encoding],axis=1)
```
