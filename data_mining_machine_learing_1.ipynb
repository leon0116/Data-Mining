```python
#2(1)
import pandas as pd
df2 = pd.read_csv('titanic.csv')
df2
```

```python
#2(3)
col=['PassengerId','Survived','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked']
for i in col:
    print(i ,'這欄有',len(df2[df2[i].isnull()].index),'個空缺值')
```

```python
#2(4)
df3=df2.drop('Cabin',axis=1)
df3
```

```python
#2(5)
df3[df3['Age'].isnull()]
```

```python
#2(6)
import numpy as np
import pandas as pd
X = df3.iloc[:,5:6].values
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan, strategy = 'median')
imp = imp.fit(X)
df3.iloc[:,5:6] = imp.transform(X)
df3
```

```python
#2(7)
import numpy as np
import pandas as pd
X = df3.iloc[:,10:11].values
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')
imp = imp.fit(X)
df3.iloc[:,10:11] = imp.transform(X)
df3
```
