```python
boston = datasets.load_boston()
x = pd.DataFrame(boston['data'], columns=boston['feature_names'])
y = pd.DataFrame(boston['target'], columns=['MEDV'])
boston_data = pd.concat([x,y], axis=1)
boston_data.head(5)
```

```python
col=['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD',
       'TAX', 'PTRATIO', 'B', 'LSTAT']
for i in col:
    print(i ,'這欄有',len(boston_data[boston_data['CRIM'].isnull()].index),'個空缺值')
```

```python
round(boston_data.corr(),2)
```

```python
from sklearn.model_selection import train_test_split
X = boston_data[['RM','LSTAT']]
y = boston_data[['MEDV']]
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=10) 
X_train.head(5)
```

```python
from sklearn import linear_model
plt.style.use('ggplot')

# linear regression 物件
regr = linear_model.LinearRegression()

# 以training datasets的資料建立線性迴歸模型
regr.fit(X_train, y_train)

# 計算R2
r_squared = regr.score(X_train, y_train)

print('截距項', regr.intercept_)
print('X係數', regr.coef_)
print('R平方:',r_squared)
```

```python
from sklearn.metrics import mean_squared_error
rmse=round(mean_squared_error(regr.predict(X_test), y_test, squared=False),4)
print("均方根誤差(RMSE):" , rmse)
```

```python
from sklearn.metrics import mean_absolute_error
mae=round(mean_absolute_error(regr.predict(X_test),  y_test),4)
print("平均絕對誤差(MAE):" , round(mae,4))
```

```python
rmse_list=[]
X = boston_data[['RM','LSTAT']]
y = boston_data[['MEDV']]
for i in range(1,100):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = i*0.01, random_state=10)
    # linear regression 物件
    regr = linear_model.LinearRegression()
    # 以training datasets的資料建立線性迴歸模型
    regr.fit(X_train, y_train)    
    rmse=round(mean_squared_error(regr.predict(X_test), y_test, squared=False),4)
    rmse_list.append(rmse)
print('均方根誤差RMSE值最小值為',min(rmse_list))
print('此時i =',rmse_list.index(min(rmse_list))+1)
```

```python
r_squared_list=[]
X = boston_data[['RM','LSTAT']]
y = boston_data[['MEDV']]
for i in range(1,100):
    X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = i*0.01, random_state=10)
    # linear regression 物件
    regr = linear_model.LinearRegression()
    # 以training datasets的資料建立線性迴歸模型
    regr.fit(X_train, y_train)
    r_squared = round(regr.score(X_train, y_train),4)
    r_squared_list.append(r_squared)
print('R平方最大值為',max(r_squared_list))
print('此時i =',r_squared_list.index(max(r_squared_list))+1)
```
